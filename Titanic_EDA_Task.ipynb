```python
# Data Handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set default theme
sns.set(style="whitegrid")

# To display plots inside the notebook
%matplotlib inline



```

```python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display the first 5 rows
titanic.head()
```

```python
# Basic info of dataset
titanic.info()

# Summary statistics for numeric and object columns
titanic.describe(include='all')

# Check for missing values
titanic.isnull().sum()
```

```python
# Check distribution of categorical columns
print(titanic['sex'].value_counts())
print(titanic['class'].value_counts())
print(titanic['embarked'].value_counts())
```

```python
# Select only numeric columns for correlation
numeric_data = titanic.select_dtypes(include=['float64', 'int64'])

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']].dropna(), hue='survived')
plt.show()
```

```python
sns.histplot(titanic['age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


sns.boxplot(x='class', y='fare', data=titanic)
plt.title("Fare by Class")
plt.show()


sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival by Gender")
plt.show()


sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title("Age vs Fare by Survival")
plt.show()


```

## Summary of Findings

- Females had a higher survival rate than males.
- Passengers in 1st class had the highest survival rate.
- Most passengers were aged between 20 and 40.
- Fare was positively correlated with survival.
- Missing values in 'age', 'embarked', and 'deck' should be handled before model training.
- Class, Sex, and Fare are important features for predicting survival.
